```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define categories in tabular form
categories = [
    {"Category Name": "Sub-Subatomic", "Category Dimension (meters)": 1e-20, "Lower Bound": 10**-22.5, "Upper Bound": 10**-17.5},
    {"Category Name": "Subatomic", "Category Dimension (meters)": 1e-15, "Lower Bound": 10**-17.5, "Upper Bound": 10**-12.5},
    {"Category Name": "Atomic", "Category Dimension (meters)": 1e-10, "Lower Bound": 10**-12.5, "Upper Bound": 10**-7.5},
    {"Category Name": "Microscopic", "Category Dimension (meters)": 1e-5, "Lower Bound": 10**-7.5, "Upper Bound": 10**-2.5},
    {"Category Name": "Human Scale", "Category Dimension (meters)": 1e0, "Lower Bound": 10**-2.5, "Upper Bound": 10**2.5},
    {"Category Name": "Planetary", "Category Dimension (meters)": 1e5, "Lower Bound": 10**2.5, "Upper Bound": 10**7.5},
    {"Category Name": "Stellar", "Category Dimension (meters)": 1e10, "Lower Bound": 10**7.5, "Upper Bound": 10**12.5},
    {"Category Name": "Solar System", "Category Dimension (meters)": 1e15, "Lower Bound": 10**12.5, "Upper Bound": 10**17.5},
    {"Category Name": "Galactic", "Category Dimension (meters)": 1e20, "Lower Bound": 10**17.5, "Upper Bound": 10**22.5},
    {"Category Name": "Cosmic Structure", "Category Dimension (meters)": 1e25, "Lower Bound": 10**22.5, "Upper Bound": 10**27.5}
]

# Define example objects in tabular form
examples = [
    {"Example Object": "Hypothetical particles (preons)", "Example Object Dimension (meters)": 1e-18},
    {"Example Object": "Proton", "Example Object Dimension (meters)": 1.6e-15},
    {"Example Object": "Neutron", "Example Object Dimension (meters)": 1.6e-15},
    {"Example Object": "Helium atom", "Example Object Dimension (meters)": 1e-10},
    {"Example Object": "Water molecule", "Example Object Dimension (meters)": 2.75e-10},
    {"Example Object": "DNA molecule", "Example Object Dimension (meters)": 2.5e-9},
    {"Example Object": "Human cell", "Example Object Dimension (meters)": 1e-5},
    {"Example Object": "Bacterium", "Example Object Dimension (meters)": 1e-6},
    {"Example Object": "Virus", "Example Object Dimension (meters)": 1e-7},
    {"Example Object": "Red blood cell", "Example Object Dimension (meters)": 8e-6},
    {"Example Object": "Human", "Example Object Dimension (meters)": 1.7},
    {"Example Object": "Building", "Example Object Dimension (meters)": 10},
    {"Example Object": "Tree", "Example Object Dimension (meters)": 10},
    {"Example Object": "Mount Everest", "Example Object Dimension (meters)": 8848},
    {"Example Object": "Earths diameter", "Example Object Dimension (meters)": 1.27e7},
    {"Example Object": "Moons diameter", "Example Object Dimension (meters)": 3.47e6},
    {"Example Object": "Distance from Earth to Moon", "Example Object Dimension (meters)": 3.84e8},
    {"Example Object": "Suns diameter", "Example Object Dimension (meters)": 1.39e9},
    {"Example Object": "Red giant star", "Example Object Dimension (meters)": 1e10},
    {"Example Object": "Supergiant star", "Example Object Dimension (meters)": 1e11},
    {"Example Object": "Distance from Earth to Sun", "Example Object Dimension (meters)": 1.5e11},
    {"Example Object": "Solar Systems diameter (Neptunes orbit)", "Example Object Dimension (meters)": 9.5e12},
    {"Example Object": "Oort Cloud", "Example Object Dimension (meters)": 1.5e15},
    {"Example Object": "Milky Way Galaxy", "Example Object Dimension (meters)": 1e21},
    {"Example Object": "Andromeda Galaxy", "Example Object Dimension (meters)": 2e21},
    {"Example Object": "Local Group of Galaxies", "Example Object Dimension (meters)": 5e22},
    {"Example Object": "Observable Universe", "Example Object Dimension (meters)": 8.8e26},
    {"Example Object": "Laniakea Supercluster", "Example Object Dimension (meters)": 5e24}
]

# Convert categories and examples to DataFrames
categories_df = pd.DataFrame(categories)
examples_df = pd.DataFrame(examples)

# Sorting function to assign examples to correct categories
def assign_category(row):
    for _, cat in categories_df.iterrows():
        if cat['Lower Bound'] <= row['Example Object Dimension (meters)'] <= cat['Upper Bound']:
            return cat['Category Name']
    return None

# Assign categories to examples
examples_df['Category Name'] = examples_df.apply(assign_category, axis=1)

# Print the category table of examples
category_table = examples_df.groupby('Category Name').apply(lambda x: x[['Example Object', 'Example Object Dimension (meters)']])
print(category_table)

# Plotting
plt.figure(figsize=(14, 8))
plt.scatter(examples_df['Category Name'], examples_df['Example Object Dimension (meters)'], color='b', marker='o', label='Example Objects')

# Plot category bounds
for _, row in categories_df.iterrows():
    plt.axhline(y=row['Lower Bound'], color='gray', linestyle='--')
    plt.axhline(y=row['Upper Bound'], color='gray', linestyle='--')
    plt.fill_between([row['Category Name']], row['Lower Bound'], row['Upper Bound'], color='gray', alpha=0.2)

plt.yscale('log')
plt.xlabel('Category Name')
plt.ylabel('Dimension (meters)')
plt.title('Object Dimensions in Log-Log Scale with Category Bounds')
plt.xticks(rotation=45)

# Adding labels for each point
for i, row in examples_df.iterrows():
    plt.annotate(row['Example Object'], (row['Category Name'], row['Example Object Dimension (meters)']),
                 textcoords="offset points", xytext=(0,10), ha='center')

plt.grid(True, which="both", ls="--")
plt.legend()
plt.tight_layout()
plt.show()
```
